```python
import pandas as pd
import matplotlib.pyplot as plt
```

We are first going to display the Nusrat taxonomy types.

```python
df = pd.read_csv('spec_generation_output.csv')
taxonomic_cts = df['taxonomy_type'].value_counts() /  len(df['taxonomy_type'])

print(taxonomic_cts)
# Step 2: Plot as a bar chart
plt.figure(figsize=(12,6))
taxonomic_cts.plot(kind='bar')

plt.title('Relative Frequency of Query Taxonomic Type')
plt.xlabel('Query Taxonomic Classification')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

Next, we can do the chart types. 

```python
chart_types = df['chart_type'].value_counts() / len(df['chart_type'])

plt.figure(figsize=(12,6))
chart_types.plot(kind='bar')

plt.title('Relative Frequency of Visualization Type')
plt.xlabel('Visualization Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

The chart types are rather skewed toward multiview, so here is the data outside of multiview. 

```python
chart_types = df['chart_type'].value_counts() /  len(df['chart_type'])

# Exclude 'multiview'
chart_types = chart_types[chart_types.index != 'multiview']

plt.figure(figsize=(12,6))
chart_types.plot(kind='bar')

plt.title('Relative Frequency of Visualization Type, Excluding Multiview Data')
plt.xlabel('Visualization Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
